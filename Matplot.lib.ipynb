```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='X Squared')
axes.plot(x,x**3,label='X cubed')
axes.legend(loc=10)  #loc stands for location code of legends(10 represends centre)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='X Squared')
axes.plot(x,x**3,label='X cubed')
axes.legend(loc=(0.1,0.7))  #loc stands for location code of legends,here we specify x and y co-ordinate
```
