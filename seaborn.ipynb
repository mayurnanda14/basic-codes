# Distribution plot

```python
import seaborn as sns
%matplotlib inline         #we can see visualizations inside the notebook
```

```python
tips=sns.load_dataset('tips') #to load inbuilt dataset of seaborn library
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill']) #will give histogram of one variable(kernal density estimation of one variable)
```

```python
sns.distplot(tips['total_bill'],kde=False) #to remove KDE
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=100)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips) #it will give histogramm as well as scatter plot
#scatter plot indicated both tip and total bill are directly proportional
#Joint plot is used for bivariate analysis i.e to find relation between two plots
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#instead of dots will give hexagons in scatter plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
#Seaborn will draw regression line on the scatter plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
#seaborn will give you density wheter the values are more
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pairplot(tips) #pair plot will do the above operation for every single variable of tips
#nice way to quickly visualize the data and get relation between all variables
#these variables will be numerical variables
```

```python
sns.pairplot(tips,hue='sex') #will divide plots in different unique values of categorical variable
#In this xample it is divided into two sex using two different colors
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
#give dash line for every value in this column
#It is used for univariate distribution
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,rug=True)
#will include rugplot in distribution plot
```

# Categorical plots

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#will give average or mean of total_bill for both category of sex
#Average total bill for males is slightly more than average bill for females
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
#insted of mean for each category it will give standard deviation
```

```python
sns.countplot(x='sex',data=tips)
#it is almost same as barplot, only difference is ,instead of mean or standard deviation it will give u count of categorical variables
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

depending upon audience, we can either use boxplot or violin plot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
#will give you the box plot
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
#will give you the violin plot
```

# Grid

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris=sns.load_dataset('iris')
sns.set(style="darkgrid")
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```
